```python
import os 
from tinydas.utils import load_das_file
import numpy as np

path = "./data"
n = 1000
```

```python
def are_all_numbers(arr):
    # Check if the array's data type is a subtype of numeric types (int or float)
    is_numeric_dtype = np.issubdtype(arr.dtype, np.number)
    
    # Check if all elements are finite numbers (not NaN or infinite)
    are_all_finite = np.all(np.isfinite(arr))
    
    return is_numeric_dtype and are_all_finite
```

```python

filenames = [entry.path for entry in os.scandir(path)]
if n != -1:
    filenames = filenames[:n]

#with ThreadPoolExecutor() as executor:
results = list(map(load_das_file, filenames))
#results = [load_das_file(fs) for fs in filenames]
results = [tup[0] for tup in results if tup[0].shape == (625, 2137)]
```

```python
count = 0
for r in results:
    if (res := are_all_numbers(r)): count += 1
count
```
